```python
from sympy import *
init_printing(use_latex="mathjax")
from IPython.display import display as print
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({"xtick.top": True , "ytick.right": True,
                     "xtick.minor.visible": True, "ytick.minor.visible": True,
                     "xtick.direction": "in" , "ytick.direction": "in",
                     "axes.labelsize": "large", "text.usetex": True, "font.size": 13
                     })
```

```python
a = symbols("a")

a1 = Matrix([a, 0, 0])
a2 = Matrix([0, a, 0])
a3 = Matrix([a/2, a/2, a/2])

V = a1.dot(a2.cross(a3))

b1 = 2 * pi * (a2.cross(a3)) / V
b2 = 2 * pi * (a3.cross(a1)) / V
b3 = 2 * pi * (a1.cross(a2)) / V

print(simplify(b1))
print(simplify(b2))
print(simplify(b3))
```

```python
r,R = symbols("r R", real=True)
k = symbols("k", positive=True)

f = r*sin(k*r)/k
g = f.integrate((r,0,R)).subs({R:1})
h = lambdify(k, g, "numpy")

X = np.linspace(0,20,1000)
Y = h(X)

plt.axhline(0,linestyle="--",c="black",alpha=0.5)
plt.plot(X,Y)
plt.xlim(0,20)
plt.xlabel("$kR$ in 1")
plt.ylabel(r"$\rho$ in $4\pi\rho_0R^3$")
plt.title("Formfaktor einer homogen geladene Kugel")
plt.savefig("formfaktor.pdf")
```
