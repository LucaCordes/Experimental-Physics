```python
from matplotlib.pyplot import *
from numpy import *

lim = [-15,5,0,0.5]

y = floor(random.normal(-5, 3, 1000)) + random.normal(0, 0.1, 1000)
y = y[y>=-15]
y = y[y<=5]
hist(y, density=True, bins=100, range=[-15,5]) # label=s"$$"

xlabel("q")
ylabel(r"Wahrscheinlichkeitsdichte für $n(q)$ ; bins=100")
xlim(lim[:2])
ylim(lim[2:])
xticks(linspace(-15,5,21))
grid()

savefig('hist.png', dpi=500)
show()
```
